```python
import pandas as pd
import bs4
import requests
```

```python
page = 1
title_list = []
desc_list = []
tag_list = []
author_list = []
while page <= 10:
    data = requests.get(f'https://www.thebangkokinsight.com/page/{str(page)}/?s=7-eleven')
    soup = bs4.BeautifulSoup(data.text)
    for n in soup.find_all('div', {'class':'list-item'}):
        title_list.append(n.find('h3',{'class':'post__title typescale-2'}).find('a').text)
        desc_list.append(n.find('div',{'class':'post__excerpt'}).find('p').text)
        tag_list.append(n.find('a',{'class':'post__cat cat-theme'}).text)
        author_list.append(n.find('span',{'class':'entry-author'}).find('a').text)
    page += 1
    print('complete page number: ',page)
```

```python
df = pd.DataFrame([title_list, desc_list, tag_list, author_list]).transpose()
df.columns = ['title','describe','tag','author']
df.set_index('title')
```

```python
df.to_csv('7eleven_news.csv',encoding='utf-8-sig')
```
